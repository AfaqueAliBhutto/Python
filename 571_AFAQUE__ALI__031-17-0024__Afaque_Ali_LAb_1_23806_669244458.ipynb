**Develop a BMI function in python programming language that gets the weight and height as
inputs and show that whether input user is underweight, normal weight, overweight or obese.**

```python
# BMI function
def BMI(height, weight):
    height_m = height/3.281
    BM = weight/(height_m**2)
    return BM
       
       
# function for catefory determination
def BMI_category(bmi):
    if(bmi < 18.5):
        return "UnderWeight"
    if(bmi >= 18.5) and (bmi < 24.9) :
        return "Normal"
    if(bmi >= 25) and (bmi < 29.9):
        return "Overweight"
    if(bmi >= 30):
        return "Obese"

print('Input Weight in Kg: ') 
w = float(input())
print('Input Hieght in Meters: ') 
h = float(input())

print('\nYour BMI is: ',BMI(h,w))
print('\nYour BMI Category is: ',BMI_category(BMI(h,w)))
```

** Write a function that takes input a number, which is temperature in degree Celsius and return the

```python
print('input in Celcius: ')
C = float(input())

F = (C * 9/5) + 32
K = C + 273.15

print('\nFahrenheit: ',F)
print('\nKelvin: ',K)
```

**Create a list of even numbers and odd numbers with the help of for loop and append function.
Than combine both lists together**

```python

EvenList = []
OddList = []
Even_odd_List = []

for even in range(1, 100 + 1): 
    
    # creating even list 
    if even % 2 == 0: 
        print(even, end = " ")
        EvenList.append(even)
for odd in range(1, 100 + 1):
    
    # creating odd list 
    if odd % 2 != 0: 
        print(odd, end = " ")
        OddList.append(odd)

# Combining both lists
Even_odd_List.extend(OddList)

Even_odd_List.extend(EvenList)
```
